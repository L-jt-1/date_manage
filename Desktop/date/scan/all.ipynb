```python
"""実行手順
データの全処理
    1.すべてのファイルを結合し、output_fileに書き込む
    2.output_file内のstopwordsを削除
    3.名詞を単数化にする
    4.動詞を原型に戻す
    
    ５.品詞ごとに単語の出現回数を数える

これからやること 
1.英語の翻訳「辞書処理(未知)」
2.アプリ作る
3.コーパスデータを増やす
"""
```

```python
#すべてのファイルを結合し、output_fileに書き込み。
input_file = ["T1-1-all.txt", "T1-2-all.txt"]
output_file = "out_file"

for file in input_file:
    with open(file, "r", encoding="utf-8") as f:
        f_read = f.read().lower()#小文字に変更する
        #print(f_read)
        
    with open(output_file, "a", encoding="utf-8") as f:
        print(f_read, file=f)

    
```

```python
#文をtokenization化にして、stopwordを削除
from nltk.corpus import stopwords
import re

with open(output_file, encoding="utf-8") as f:
    f_read = f.read()
    #分かち書き
    str_nopunc = re.findall(r'\b\w+\b', f_read)#すべての読点が削除でき、リストになる
    #print(str_nopunc)#各単語が入るリストになる['she', 's', 'arranging', 'some', 'flowers', ....]
    #stopwordsを削除
    no_stopwords = [word for word in str_nopunc if word not in stopwords.words('english')]#list
    # for word in no_stopwords:
    #     if len(word) != 1 or type(word) is int:
    #         w_index = no_stopwords.index(word)
    #         no_stopwords.pop(w_index)
    print(no_stopwords)
    #print(str_nopunc)
```

```python
#名詞複数を単数に変換lemmatize
from nltk import stem
#nltk.download('wordnet')

wnl = stem.wordnet.WordNetLemmatizer()
print(wnl)
no_punc_s = []
new_no_punc_s = []
for word in no_stopwords:
    w = wnl.lemmatize(word)#単数化
    
    if len(w) != 1:#アルファベットを削除,文字列中の数字まだ削除してない！
        no_punc_s.append(w)
        no_punc_no_s = re.sub(r'[0-9]+', '', w)#数字を削除　
        new_no_punc_s.append(no_punc_no_s)
print(new_no_punc_s)
```

```python
#品詞の取得pos_tag,

import nltk
#tags = []
#for word in new_no_punc_s:
   #print(word)#アルファベット、数字、stopwordsが入ってないただ文字列の各単語arranging flower moving

tags = nltk.pos_tag([word for word in new_no_punc_s if word])#空文字列だとダメ！
    #tags.append(tag)
#print(type(new_no_punc_s))    
print(tags)
#結果：[('arranging', 'VBG'), ('flowers', 'NNS'),.....］
    
```

```python
#動詞を原型に変更

# 動詞（VB原型，VBD过去式，VBG动名词或者现在分词，VBN过去分词, 
# VBP動詞 (三人称単数以外の現在形), VBZ動詞 (三人称単数の現在形)）
# を原型(VB)に変更
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()

new_tags = []

for s in range(len(tags)):
    word = tags[s][0]
    word_type = tags[s][1]
    if word_type == "VBD" or word_type == "VBG" or word_type == "VBN" or word_type == "VBZ" or word_type == "VBP":
        word_type = wnl.lemmatize(word, 'v')
        new_tags.append((word_type, "VB"))
    else:
        new_tags.append(tags[s])
print(new_tags)    
```

```python
#二次リスト内の各要素に同じ品詞な単語を入れる

pos_list = []#品詞のみ
word_pos_list = []#二次リスト

for i in range(len(new_tags)):
    pos_list.append(new_tags[i][1])
pos_num = len(set(pos_list))#品詞の種類を計算
p_list = list(set(pos_list))#品詞種類のリストを作る

for p in range(len(p_list)):#品詞数の空リストを作成
    word_pos_list.append([])
    for i in new_tags:
        if i[1] == p_list[p]:
            word_pos_list[p].append(i)
            
print(word_pos_list)
print(len(word_pos_list))#23種類
```

```python
#単語の出現回数を数える（查找列表中的重复元素并统计重复数量）
from collections import Counter

w_p_count = []
for w_p in word_pos_list:
    w_p_count.append(Counter(w_p))
print(w_p_count)
print(len(w_p_count))
```

```python
#品詞ごとをリストで分類する
n_type = ["NNP", "NN", "PRP", "WP", "WP$", "NNS"]
v_type = ["VB"]
adj_type = ["JJ", "JJR", "JJS"]
adv_type = ["RB", "RBR", "RBS", "WRB"]
i_type = ["IN"]
o_type = ["PDT", "CC", "CD", "FW", "DT", "RP", "UH", "MD"]

n = []
v = []
adj = []
adv = []
i = []
o = []

for count in w_p_count:
    for word_and_type in count:
        if word_and_type[1] in n_type:
            n.append(word_and_type)
        elif word_and_type[1] in v_type:
            v.append(word_and_type)
        elif word_and_type[1] in adj_type:
            adj.append(word_and_type)
        elif word_and_type[1] in adv_type:
            adv.append(word_and_type)
        elif word_and_type[1] in i_type:
            i.append(word_and_type)
        else:
            o.append(word_and_type)
#print(n, len(n))           
print(o, len(o))
            
```

```python
import sys 
import urllib.request
import urllib.parse
from html.parser import HTMLParser

WEBLIO_URL = "http://ejje.weblio.jp/content/"

class MyHTMLParser(HTMLParser):
    def __init__(self):
        HTMLParser.__init__(self)
        self._is_midashigo = False
        self._is_translated_word = False

    def handle_starttag(self, tag, attrs):
        if tag == "h2":
            attrs_dict = {t[0]:t[1] for t in attrs}
            if attrs_dict.get("class", None) == "midashigo":
                self._is_midashigo = True
        if tag == "a":
            attrs_dict = {t[0]:t[1] for t in attrs}
            if attrs_dict.get("class", None) == "crosslink" and self._is_midashigo:
                self._is_translated_word= True

    def handle_data(self, data):
        if self._is_translated_word:
            print(data)
            self._is_midashigo = False
            self._is_translated_word = False


def parse_weblio(raw_ret):
    parser = MyHTMLParser()
    parser.feed(raw_ret)


def main():
    word = urllib.parse.quote_plus(sys.argv[1])
    filename, headers = urllib.request.urlretrieve(WEBLIO_URL + word)
    with open(filename, encoding="utf-8") as f:
        parse_weblio(f.read())



if __name__ == "__main__":
    main()
```
